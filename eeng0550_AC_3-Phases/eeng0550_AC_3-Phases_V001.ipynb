```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi, cos, sin, tan, sqrt, ones
```

```python
def AC(t, A=1., f=1., phi=0.):
    omega = 2*pi*f
    return(A*cos(omega*t + phi))
```

```python
t = np.linspace(0,0.1,1000)

fig1, ax1 = plt.subplots(figsize=(8,4), dpi = 136)
Vrms = 230.
V0 = Vrms*sqrt(2.)
f = 50.
ax1.plot(t,AC(t, V0, f), label=r"$\cos$")
ax1.plot(t,AC(t, V0, f, phi=-pi/2), label=r"$\sin$")
ax1.plot(t, ones(len(t))*Vrms, "g--", label=r"$V_\mathrm{rms}$")
ax1.plot(t, -ones(len(t))*Vrms, "g--")
ax1.set_xlabel(r"Time $t$ / s")
ax1.set_ylabel(r"Voltage $V$ / V")
ax1.set_xlim(0,0.1)
ax1.legend()
ax1.grid(True)
```

Integrals of $\sin$ and $\cos$ over one period yield 0 because areas above 0 are equal to area below 0 for a single period:

$$
\int_0^T A\sin(\omega t)\,dt = 0
$$

$$
\int_0^T A\cos(\omega t)\,dt = 0
$$

Integrals of $\sin^2$ and $cos^2$ over one period are positive and equal because of symmetry:

$$
\int_0^T \left(A\cos(\omega t)\right)^2\,dt = \int_0^T \left(A\sin(\omega t)\right)^2\,dt
$$

$$
A^2 \int_0^T \cos^2(\omega t)\,dt = A^2 \int_0^T \sin^2(\omega t)\,dt
$$

This general identity also holds for $A=1$:

$$
\int_0^T \cos(\omega t)^2\,dt = \int_0^T \sin(\omega t)^2\,dt
$$

Another piece of information: Pythagoras in a unit circle yields for all angles $\phi$:

$$
\cos^2 \phi + \sin^2 \phi = 1 \, , \, \mathrm{for\;all}\;\phi
$$

THe angle $\phi$ can be even changing in time, i.e. $\phi = \omega t$. Pythagoras still holds:

$$
\cos^2 (\omega t) + \sin^2 (\omega t)= 1
$$

Let us integrate both sides of the equation over one period:

$$
\int_0^T \cos^2 (\omega t)\, dt + \int_0^T \sin^2 (\omega t)\, dt = \int_0^T 1\, dt = T
$$

From symmetry we can infer that the integrals 

$$
2\,\int_0^T \cos^2 (\omega t)\, dt = T
$$

Finally we get:

$$
\int_0^T \cos^2 (\omega t)\, dt = \int_0^T \sin^2 (\omega t)\, dt = \frac{T}{2}
$$


### RMS: Root of Mean Square

$$
f(t) = A\cos(\omega t) \\ \left( \mathrm{or} \, f(t) = A\sin(\omega t) \right)
$$

Square: 

$$
f^2(t) = A^2 \cos^2(\omega t)
$$

Mean of the squares:

$$
\left< f^2 \right>  = \frac{1}{T} \int_0^T f^2(t) dt = \frac{1}{T} \int_0^T A^2 \cos^2(\omega t) dt = A^2 \frac{1}{T} \int_0^T \cos^2(\omega t) dt = \frac{A^2}{2} 
$$

The root of the mean of the squares:


$$
f_\mathrm{rms} = \sqrt{\left< f^2 \right>} =  \frac{A}{\sqrt{2}} 
$$

Summary:

$$
V_\mathrm{rms}\left< f^2 \right>
= \sqrt{ \frac{1}{T} \int_0^T \left( V(t) \right)^2 dt}  
= \sqrt{ \frac{1}{T} \int_0^T V^2(t) dt}  
= \sqrt{ \frac{1}{T} \int_0^T \left( V_0 \cos(\omega t) \right)^2 dt} =  \frac{V_0}{\sqrt{2}} 
$$

## Yet Another Graphical Interpretation of RMS

The following harmonic (aka (co-)sinusoidal) function with amplitude $V_0$ and freequency $f = 1/T$  is given:

$$
V(t) = V_0\,\cos\left(\frac{2\pi}{T}\,t\right)
$$

The square of the function is:

$$
V^2(t) = V^2_0\,\cos^2\left(\frac{2\pi}{T}\,t\right)
$$

The figure below shows $V^2(t)$ with an example amplitude $V_0 = 2$.

```python
fig3, ax3 = plt.subplots(figsize=(6,3), dpi = 136)

T = 1. # 
f = 1/T
A = 2.

t = np.linspace(0.,T,1000)

x = t/T # time in units of T
y = AC(t, A, f)**2

ax3.plot(x, y, linestyle="-", color = "tab:blue", label=r"$V\,^2(t) = V_0^2\,\cos^2\left(\frac{2\pi}{T}\,t\right)$")
#ax3.plot(x, np.ones(x.shape)*4, linestyle="-.", color = "black", label=r"$V_0^2 = 4$")
ax3.plot(x, np.ones(x.shape)*4, linestyle="-.", color = "black")
ax3.plot(x, np.ones(x.shape)*0, linestyle="-", color = "black")

ax3.fill_between(x, y, 0, color='tab:blue',   alpha=0.1)
ax3.fill_between(x, y, 4, color='tab:red', alpha=0.1)

ax3.annotate(r"$V_0^2 = 4$",(0.25,4.1))

ax3.grid(True)
ax3.set_xlabel(r"Time $t$ in T")
ax3.set_ylabel(r"$V\,^2(t)$")
ax3.set_xlim(-0.05,1.05)
ax3.set_ylim(-.5,4.5)
ax3.legend()
plt.plot()
```

The rectangular area of the total shaded region if the figure abobe is: 
$$
A   = T V_0^2
$$

The total area is the sum of the blue and the red areas:

$$
A = A_\mathrm{red} + A_\mathrm{blue}
$$

From the symmetry follows:

$$
A_\mathrm{red} = A_\mathrm{blue} 
$$

Thus the blue area yields:

$$
A_\mathrm{blue} = \frac{A}{2} = \frac{T V_0^2}{2}
$$

The blue area is the integral of the the function $V^2(t)$ over the interval $[0 \ldots T]$:

$$
A_\mathrm{blue} = \int_0^T V^2(t)dt = \frac{T V_0^2}{2}
$$

The mean amplitude of $V^2(t)$ (mean square) over the interval $[0 \ldots T]$ is:

$$
\left<V^2\right>  = \frac{1}{T}\int_0^T V^2(t)\,dt = \frac{A_\mathrm{blue}}{T} = \frac{1}{T} \cdot \frac{T V_0^2}{2} = \frac{V_0^2}{2}
$$


```python
fig3, ax3 = plt.subplots(figsize=(6,3), dpi = 136)

T = 1. # 
f = 1/T
A = 2.

t = np.linspace(0.,T,1000)

x = t/T # time in units of T
y = AC(t, A, f)**2

ax3.plot(x, y, linestyle="-", color = "tab:blue")

#ax3.plot(x, y, linestyle="-", color = "tab:blue", label=r"$V\,^2(t) = V_0^2\,\cos^2\left(\frac{2\pi}{T}\,t\right)$")
#ax3.plot(x, np.ones(x.shape)*4, linestyle="-.", color = "black", label=r"$V_0^2 = 4$")

ax3.plot(x, np.ones(x.shape)*2, linestyle="--", color = "black")
#ax3.plot(x, np.ones(x.shape)*0, linestyle="-", color = "black")

ax3.fill_between(x, y, 0, color='tab:blue',   alpha=0.1)
#ax3.fill_between(x, y, 4, color='tab:red', alpha=0.1)

#ax3.annotate(r"$\left<V^2\right> = V^2_0/2 = 2$",(0.13,2.15))

ax3.annotate(r"$\left<V^2\right> = \frac{1}{T}\int_0^T V^2(t)\, dt = V_0^2 / 2$",(0.55,3), bbox={"alpha": 0.7,  "boxstyle": "square", "facecolor": "lightgrey", "edgecolor": "blue"})
ax3.annotate(r"$\left<V^2\right>$",(0.22,2.15))

ax3.set_title(r"Mean Square, $V_0=2$")

ax3.grid(True)
ax3.set_xlabel(r"Time $t$ in T")
ax3.set_ylabel(r"$V\,^2(t)$")
ax3.set_xlim(-0.05,1.05)
ax3.set_ylim(-.5,4.5)
#ax3.legend()
plt.plot()
```

## Example: AC Standard in Germany

The RMS voltage in a standard (sinusoidal or harmonic) AC is called the **effective voltage**: $V_\mathrm{eff} = V_\mathrm{rms}$. 

The German domestic AC power grid has an effective voltage of $V_\mathrm{eff} = 240\,\mathrm{V}$.

The voltage amplitude is $V_0 = V_\mathrm{eff} \sqrt{2} \approx  340\,\mathrm{V}$.


$$
f_1(t) = e^{i\omega t} \\
f_2(t) = e^{i\omega t+i\phi} \\
$$

$$
f_2-f_1 = e^{i\omega t} (e^{i\phi}-1) = h\cdot e^{i\omega t}
$$

$$
h = e^{i\phi}-1 = |h|e^{i\alpha}
$$ 

$h$ is a complex number yielding a amplitude modification $|h|$ (gain, attenuation) and phase shift $\alpha$. 

$$
\mathrm{Re}(h) = \cos\phi -1 \\
\mathrm{Im}(h) = \sin\phi\\
$$

Absolute value of h:

$$
|h| = \sqrt{\mathrm{Re}^2 + \mathrm{Im}^2} = \sqrt{2}\sqrt{1-\cos\phi}
$$

Trigonometric Identity:

$$
1-\cos\phi = 2\sin^2(\phi/2)
$$

It follows:

$$
|h| = 2\,\sqrt{\sin^2(\phi/2)} = 2 \left|\sin(\phi/2)\right|
$$


Phase angle $\alpha$:

$$
\tan\alpha = \frac{\mathrm{Im}}{\mathrm{Re}}
$$

## Check the Integral Calculations for the RMS Values by Numerical Integration with Trapezoids

```python
t = np.linspace(0, T, 129) # -20ms ... 20ms , two periods

f = 50. # Hz
T = 1/f # = 20 ms

phi1 = 0.        # aka 0
phi2 = (2*pi)/3  # aka 120
phi3 = 2*phi2    # aka 240

Vrms = 230.

V0 = V10 = V20 = V30 = Vrms*sqrt(2)

L1 = V1 = AC(t, V10, f, phi1)
L2 = V2 = AC(t, V20, f, phi2)
L3 = V3 = AC(t, V30, f, phi3)

V12 = V2 - V1
V23 = V3 - V2
V31 = V1 - V3

fig, ax = plt.subplots(figsize=(8,4), dpi = 136)

ax.plot(t/T ,V12**2, "-", label=r"$(V_2 - V_1)^2$")
ax.plot(t/T ,V23**2, "-", label=r"$(V_3 - V_2)^2$")
ax.plot(t/T ,V31**2, "-", label=r"$(V_1 - V_3)^2$")

ax.legend()
ax.grid(True)
plt.plot()
```

**Numerical Integration of Squares (Sum of Squares)**

```python
# INtegrate numerically with trapezoids
Integral = np.trapz(V12**2,t)
print(f"\nIntegral of Squared Voltage V12^2 = (V2-V1)^2 between Phase 1 and 2: {Integral = :.3g} V\n")
```

**The Mean of the Integral: Mean Square**

```python
MS = Integral/T
print(f"\nMean Square: {MS = :.3g} V\n")
```

**Root Mean Square**

```python
RMS = sqrt(MS)
print(f"\nRoot Mean Square of (V2-V1) by Integration: {RMS = :.1f} V\n")
```
